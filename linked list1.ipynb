```python
"""Given a singly linked list, deleteÂ **middle**Â of the linked list. For example, if given linked list is 1->2->**3**->
4->5 then linked list should be modified to 1->2->4->5.If there areÂ **even**Â nodes, then there would beÂ **two middle**Â 
nodes, we need to delete the second middle element. For example, if given linked list is 1->2->3->4->5->6 then it should be
modified to 1->2->3->5->6.If the input linked list is NULL or has 1 node, then it should return NULL
"""
class Node:
    def __init__(self,data=None):
        self.data = data
        self.next = None

class LinkedList:
  
    def __init__(self):
        self.head = None
        
    def delete_middle(head):
        if head is None or head.next is None:
            return None
        prev = None
        slow = head
        fast = head
        
        while fast is not None and fast.next is not None:
            fast = fast.next.next
            prev = slow
            slow = slow.next
            
        prev.next = slow.next
        
        return head
    
```

```python
"""
Given a linked list ofÂ **N**Â nodes. The task is to check if theÂ linked list has a loop. Linked list can containÂ self loop.
"""
class Node:
    def __init__(self,data= None):
        self.data = data
        self.next = None

class LinkedList:
  
    def __init__(self):
        self.head = None
        
    def detect_loop(head):
        
        if head is None or head.next is None:
            return None
        
        slow = head
        fast = head
        
        while fast is not None and fast.next is not None:
            fast = fast.next.next
            slow = slow.next
            
        if slow == fast:
            return True
        
        return False
```

```python
"""Given a linked list consisting ofÂ LÂ nodes and given a numberÂ N. The task is to find theÂ NthÂ node from the end of the 
linked list."""

class Node:
    def __init__(self,data=None):
        self.data = data
        self.next = None
        
class LinkedList:
  
    def __init__(self):
        self.head = None
        
    def nth_element(head,N):
        
        if head is None or N<=0:
            return None
        
        count = 0
        ref = head
        main = head
        
        while count < N:
            if ref is None:
                return None
            
            ref = ref.next
            count += 1
            
        if ref is not None:
            main = main.next
            ref = ref.next
            
        return main.data
```

```python
"""Given a singly linked list of characters, write a function that returns true if the given list is a palindrome,
else false."""

class Node:
    def __init__(self,data=None):
        self.data = data
        self.next = None
        
class LinkedList:
  
    def __init__(self):
        self.head = None
        
    def palindrome(head):
        fast = head
        slow = head
        
        while fast != None and fast.next != None:
            slow = slow.next
            fast = fast.next.next
            
        #middele is slow
        #now reverse the second half
        prev = slow
        slow = slow.next
        prev.next = None
        
        while slow != None:
            temp = slow.next
            slow.next = prev
            prev = slow
            slow = temp
            
        fast = head
        slow = prev
        
        while slow != None:
            if slow.val != fast.val:
                return False
            slow = slow.next
            fast = fast.next
        return True
```
